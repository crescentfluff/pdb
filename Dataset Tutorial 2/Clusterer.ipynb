```python
import pandas as pd
import matplotlib.pyplot as plot
labels = pd.read_csv('labels.txt', delimiter=r"\s+", usecols=[1], names = ['lab'])
labels.head(39)
```

```python
labels.head()
```

```python
water = pd.read_csv('water-treatment.csv', names = labels['lab'])
water.shape
```

```python
for column in water:
    if column != 'prediction':
        water[column] = pd.to_numeric(water[column], errors='coerce')
    else:
        print ('non numerical')
```

```python
water.head()
```

```python
water.describe()
```

```python
water.dtypes
```

```python
for column in water:
    if column != 'prediction':
        water[column] = water[column].fillna(water[column].mean())
    else:
        print ('non numerical')
water.head()
```

```python
pred = water['prediction']
head = water.iloc[:,1:39]
head.head()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
test = SelectKBest(score_func=chi2, k=10).fit(head, pred)

feature = test.transform(head).get_support()
```
