```python
import pandas as pd
import matplotlib.pyplot as plot
import random as rd
import math
insr = pd.read_csv('insurance.csv')
insr.head(10)
```

```python
insr2 = insr
insr2['sex'] = insr2['sex'].astype('category')
insr2['smoker'] = insr2['smoker'].astype('category')
insr2['region'] = insr2['region'].astype('category')

col = insr2.select_dtypes('category').columns
insr2[col] = insr2[col].apply(lambda x : x.cat.codes)
```

```python
insr2.head()
```

```python
insr2.sample(15)
```

```python
insr2.to_csv('insurance_modif.csv', sep=",")
```

```python
insr2.plot(x='age', y='charges', kind='scatter')
plot.show
```

```python
insr2.plot(x='age', y='bmi', kind='scatter')
plot.show
```

```python
corcol = insr2[['age', 'bmi', 'charges']]

corcol.corr(method='pearson').style.format("{:.2}").background_gradient(cmap
=plot.get_cmap('coolwarm'), axis=1)
```
